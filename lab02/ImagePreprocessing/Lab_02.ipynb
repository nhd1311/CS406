```python
def calculate_histogram(image_path):    
    image = cv2.imread(image_path, cv2.IMREAD_COLOR)
    image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
    hist_r = cv2.calcHist([image], [0], None, [256], [0, 256])
    hist_g = cv2.calcHist([image], [1], None, [256], [0, 256])
    hist_b = cv2.calcHist([image], [2], None, [256], [0, 256])
    cv2.normalize(hist_r, hist_r)
    cv2.normalize(hist_g, hist_g)
    cv2.normalize(hist_b, hist_b)
    hist = np.concatenate((hist_r, hist_g, hist_b)).flatten()
    return hist
```

```python
def read_data(dir_root_path):
    hist_list = []
    path_list = []
    for dir_name in os.listdir(dir_root_path):
        dir_path = os.path.join(dir_root_path, dir_name)
        for file_name in os.listdir(dir_path):
            file_path = os.path.join(dir_path, file_name)
            path_list.append(file_path)

            print(file_path)
            hist = calculate_histogram(file_path)
            hist_list.append((file_path, hist))
    return path_list, hist_list
```

```python
path_list, hist_list = read_data(seg_path)
```

```python
import pickle

pickle_path = 'features.pkl'

with open(pickle_path, 'wb') as f:
    pickle.dump(hist_list, f)

with open(pickle_path, 'rb') as f:
    load_hist = pickle.load(f)
```

```python
def compare_hist(hist1, hist2):
    distance = np.sqrt(np.sum((hist1 - hist2) ** 2))
    return distance
```

```python
import random

def get_random_image(dir_test_path):
    path_list = []
    for dir_name in os.listdir(dir_test_path):
        dir_path = os.path.join(dir_test_path, dir_name)
        for file_name in os.listdir(dir_path):
            file_path = os.path.join(dir_path, file_name)
            path_list.append(file_path)

    random_image_path = random.choice(path_list)
    return random_image_path
```

```python
def find_similar_images(hist_input, load_hist):
    results = []
    
    for hist in load_hist:
        dir_name = hist[0]  
        hist_data = hist[1]  
        distance = compare_hist(hist_input, hist_data)  
        results.append((dir_name, distance))  
    
    results.sort(key=lambda x: x[1])
    
    return results[:10]
```
